1. Seaborn 기본 그래프의 활용

1) Scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.rand(50)
```

```python
np.arange(50)
```

2) x, y, colors, area 설정

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 250
```

```python
plt.scatter(x, y, s=area, c=colors)
```

```python
sns.scatterplot(x=x, y=y, size = area, sizes=(area.min(), area.max()), hue=area, palette='icefire')
plt.show()
```

3) cmap과 alpha 설정

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.1)
g.get_legend().remove()
plt.title('alpha=0.1')

plt.subplot(132)
plt.title('alpha=0.5')
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.5)
g.get_legend().remove()

plt.subplot(133)
plt.title('alpha=1.0')
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.9)
g.get_legend().remove()

plt.show()
```

(1) Barplot, Barhplot - 1개의 canvas 안에 다중 그래프 그리기

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

sns.barplot(x=x, y=y, alpha=0.8, palette='YlGnBu')

plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```

 - 비교 그래프 그리기

```python
sns.barplot(x='sex', y='survived', hue='pclass', data=titanic, palette="muted")
plt.show()
```

(2) Box Plot

```python
# 샘플 데이터 생성
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
sns.boxplot(data=data, orient='h', width=0.2)
plt.subplot(122)
sns.boxplot(data=data, orient='v', width=0.2)
plt.show()
```

```python
outlier_marker = dict(markerfacecolor='r', marker='o')
sns.boxplot(x='pclass', y='age', hue='survived', data=titanic, flierprops=outlier_marker)
plt.show()
```

3) 히스토 그램

```python
N = 100000
bins = 30

x = np.random.randn(N)

sns.histplot(x, bins=bins, kde=False, color='g')
# 구 버전
# sns.distplot(x, bins=bins, kde=False, color='g')

plt.show()
```

```python
sns.histplot(x, bins=bins, kde=True, color='g')
```

```python
sns.histplot(y=x, bins=bins, kde=True, color='navy')
```

- 분포표, 누적 분포표

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(12, 4)

sns.histplot(x=x, bins=bins, ax=axes[0])
sns.histplot(x=x, bins=bins, cumulative=True, ax=axes[1])

# 구버전
# sns.displot(x=x, bins=bins)
# sns.displot(x=x, bins=bins, cumulative=True)

plt.show()
```

4) Line Plot

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

sns.lineplot(x=x, y=y_1, label='1+sin', color='blue', linestyle=':')
sns.lineplot(x=x, y=y_2, label='1+cos', color='red', linestyle='-.')

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```
