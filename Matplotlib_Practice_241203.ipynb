```python
import matplotlib.pyplot as plt
import numpy as np

# 0에서 5까지
# 0.2 간격으로 균일하게 샘플된 시간
t = np.arange(0., 5., 0.2)

# 빨간 대쉬, 파란 사각형, 녹색 삼각형

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()


# x축 값: t (시간)
# y축 값: t (y = x 선)
# 스타일: 'r--'은 빨간색(r) 점선(--)입니다.

# x축 값: t
# y축 값: t**2 (제곱 값, y = x^2)
# 스타일: 'bs'는 파란색(b) 사각형(s) 마커입니다.

# x축 값: t
# y축 값: t**3 (세제곱 값, y = x^3)
# 스타일: 'g^'는 녹색(g) 위쪽 삼각형(^) 마커입니다.
```

5. Matplotlib 축 범위 지정하기

matplotlib.pyplot 모듈의 xlim(), ylim(), axis() 함수를 사용하면 그래프의 X, Y축이 표시되는 범위를 지정할 수 있습니다.

xlim() - X축이 표시되는 범위를 지정하거나 반환합니다.

ylim() - Y축이 표시되는 범위를 지정하거나 반환합니다.

axis() - X, Y축이 표시되는 범위를 지정하거나 반환합니다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])      # X축의 범위: [xmin, xmax]
plt.ylim([0, 20])     # Y축의 범위: [ymin, ymax]

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0, 5, 0, 20])  # X, Y축의 범위: [xmin, xmax, ymin, ymax]

# axis() 함수에 입력한 리스트 (또는 튜플)는 반드시 네 개의 값 (xmin, xmax, ymin, ymax)이 있어야 합니다.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
# plt.colorbar()
plt.colorbar(shrink=0.8, aspect=10)

# plt.colorbar() 함수에서 shrink와 aspect는 컬러바의 크기와 형태를 조정하는 데 사용됩니다.

# shrink 파라미터:
# 의미: 컬러바의 길이를 조정합니다.
# 값 범위: 0.0 ~ 1.0 (또는 그 이상도 가능)
# shrink=1.0: 기본 크기 (100%)
# shrink=0.8: 컬러바 길이를 80%로 축소
# shrink=1.5: 컬러바 길이를 150%로 확장

# aspect 파라미터:
# 의미: 컬러바의 너비와 길이의 비율을 설정합니다.
# aspect 값이 크면 컬러바가 길고 얇게 나타납니다.
# 값이 작으면 컬러바가 짧고 두껍게 나타납니다.
# 기본값: 20
# aspect=10: 컬러바가 더 짧고 두껍게 나타남
# aspect=30: 컬러바가 더 길고 얇게 나타남

plt.show()
```

24. 여러개의 그래프 그리기

1) 기본사용

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

2) 축 공유하기

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.ylabel('Damped oscillation')
plt.xticks(visible=False)

ax2 = plt.subplot(2, 1, 2, sharex=ax1)    # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

31. 객체 지향 인터페이스 2

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x))      # left-top
ax[0][1].plot(x, x)               # right-top
ax[1][0].plot(x, -x+5)            # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5))   # right-bottom

plt.show()
```

2) 스타일 설정하기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x), 'gray', linewidth=3)
ax[0][1].plot(x, x, 'g^-', markersize=10)
ax[1][0].plot(x, -x+5, 'ro--')
ax[1][1].plot(x, np.sqrt(-x+5), 'b.-.')

plt.show()
```

3) 제목과 범례 표시하기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x), 'gray', linewidth=3, label='y=np.sqrt(x)')
ax[0][0].set_title('Graph 1')
ax[0][0].legend()
ax[0][1].plot(x, x, 'g^-', markersize=10, label='y=x')
ax[0][1].set_title('Graph 2')
ax[0][1].legend(loc='upper left')
ax[1][0].plot(x, -x+5, 'ro--', label='y=-x+5')
ax[1][0].set_title('Graph 3')
ax[1][0].legend(loc='lower left')
ax[1][1].plot(x, np.sqrt(-x+5), 'b.-.', label='y=np.sqrt(-x+5)')
ax[1][1].set_title('Graph 4')
ax[1][1].legend(loc='upper center')

plt.show()
```
